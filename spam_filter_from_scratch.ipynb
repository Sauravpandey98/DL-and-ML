# Initialising parameters

```python
# initialise parameters
def initialise_parameters(layer_dimensions):
    parameters={}
    for i in range(1,len(layer_dimensions)):
        parameters["W"+str(i)]=np.random.randn(layer_dimensions[i],layer_dimensions[i-1])*np.sqrt(2/layer_dimensions[i-1])
        parameters["b"+str(i)]=np.zeros((layer_dimensions[i],1))
    return parameters
#print(initialise_parameters([4,3,1]))
```

## Helping functions for forward propagation

```python
def relu(Z):
    return np.maximum(0,Z)

def sigmoid(Z):
    return 1/(1+np.exp(-Z))
```

```python
def linear_and_activation_relu(A_prev,W,b):
    Z=np.dot(W,A_prev)+b
    A=relu(Z)
    linear_cache=(A_prev,W,b)
    Activation_cache=A
    cache=(linear_cache,Activation_cache)
    return A,cache
```

```python
def linear_and_activation_sigmoid(A_prev,W,b):
    Z=np.dot(W,A_prev)+b
    A=sigmoid(Z)
    linear_cache=(A_prev,W,b)
    Activation_cache=A
    cache=(linear_cache,Activation_cache)
    return A,cache
```

```python
def forward_layer_cal(A,W,b,activation):
    if activation=="relu":
        A,cache1=linear_and_activation_relu(A,W,b)
    elif activation=="sigmoid":
        A,cache1=linear_and_activation_sigmoid(A,W,b)
    return A,cache1
```

# Forward propagation

```python
 def forward_module(X,parameters):
    caches=[]
    cache=()
    A=X
    layers=int(len(parameters)/2)
    
    for i in range(1,layers):
        A_prev=A
        A,cache=forward_layer_cal(A_prev,parameters["W"+str(i)],parameters["b"+str(i)],"relu")
        caches.append(cache)
    AL,cache=forward_layer_cal(A,parameters["W"+str(layers)],parameters["b"+str(layers)],"sigmoid")
    caches.append(cache)
    return AL,caches
```

# Cost

```python
def cost_cal(Al,Y,m):
    cost=-(1/m)*(np.dot(np.log(Al),Y)+np.dot(np.log(1-Al),(1-Y)))
    cost = np.squeeze(cost)
    return cost
```

# Helping function for back propagation

```python
def linear_backward(dZ,linear_cache):
    A_prev,W,b=linear_cache
    m=A_prev.shape[1]
    dW = (1/m)*(np.dot(dZ,A_prev.T))
    db = (1/m)*np.sum(dZ,axis=1,keepdims=True)
    dA_prev = np.dot(W.T,dZ)
    return dA_prev,dW,db  
```

```python
def sigmoid_backward(dA,A):
    dZ=dA*A*(1-A)
    return dZ
def relu_backward(dA,A):
    dZ = np.array(dA, copy=True)
    dZ[A <=0]=0
    return dZ
```

```python
#finding derivative of parameters with respect to loss
def backward_layer_cal(dA,cache,activation):
    linear_cache,activation_cache=cache
    if activation=="sigmoid":
        dZ=sigmoid_backward(dA,activation_cache)
        dA,dW,db=linear_backward(dZ,linear_cache)
    elif activation=="relu":
        dZ=relu_backward(dA,activation_cache)
        dA,dW,db=linear_backward(dZ,linear_cache)
    return dA,dW,db
```

# Back propagation

```python
def backward_prop_module(cost,AL,cache,m,Y):
    grads={}
    layers=len(cache)
    #print(type(Y))
    #print(AL.shape)
    Y=Y.reshape(AL.shape)
    
    dAL= -np.divide(Y,AL)+np.divide(1-Y,1-AL)
    #print(dAL.shape)
    
    sigmoid_cache=cache[layers-1]
    
    
    grads["dA"+str(layers-1)],grads["dW"+str(layers)],grads["db"+str(layers)]=backward_layer_cal(dAL,sigmoid_cache,"sigmoid")
    #print(grads)
    
    for l in reversed(range(layers-1)):
        #print(l)
        current_cache=cache[l]
        grads["dA"+str(l)],grads["dW"+str(l+1)],grads["db"+str(l+1)]=backward_layer_cal(grads["dA"+str(l+1)],current_cache,"relu")
        
        
    return grads  
    
    
    
```

# Updating parameters

```python
def update_parameters(grads,parameters,learning_rate,layers):
    for l in range(1,layers+1):
        parameters["W"+str(l)]=parameters["W"+str(l)]-learning_rate*grads["dW"+str(l)]
        parameters["b"+str(l)]=parameters["b"+str(l)]-learning_rate*grads["db"+str(l)]
    return parameters
```

# L-layer deep learning model

```python
def deep_learning_module(X,Y,learning_rate,layer_dim,iteration):
    m=X.shape[1]
    layers=len(layer_dim)-1
    parameters=initialise_parameters(layer_dim)
    for i in range(iteration):
        AL,cache=forward_module(X,parameters)
        Cost=cost_cal(AL,Y,m)
        print("cost at"+str(i)+" iteration:"+" "+"{}".format(Cost))
        grads=backward_prop_module(Cost,AL,cache,m,Y)
        parameters=update_parameters(grads,parameters,learning_rate,layers)
        
    return parameters
    
    
```

```python
def predict(X_test,learned_parameters,Y):
    AL,cache=forward_module(X_test,learned_parameters)
    prediction=(AL>=0.5).astype(int)
    prediction=prediction.reshape(Y.shape)
    accuracy = accuracy_score(Y,prediction,normalize=True)
    return accuracy
    
```

```python
learned_parameters=deep_learning_module(X,Y,0.1,[3000,100,10,1],1500)
print("train_set accuracy is={}".format(predict(X,learned_parameters,Y)))

print("test_set accuracy is={}".format(predict(devX_test,learned_parameters,devY_test)))
```
